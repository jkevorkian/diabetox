```python
import pandas as pd

# -----------------------------
# Input / Output paths
# -----------------------------
INPUT_CSV = "predictions.csv"       
OUTPUT_CSV = "predictions_clean.csv"

# -----------------------------
# Load the dataset
# -----------------------------
df = pd.read_csv(INPUT_CSV)

# -----------------------------
# Ensure patient_id is in the format patient_XXXXX
# -----------------------------
def format_patient_id(x):
    x_str = str(x)
    # Remove existing 'patient_id' or 'patient_' prefix
    if x_str.startswith("patient_id"):
        numeric_part = x_str.replace("patient_id", "")
    elif x_str.startswith("patient_"):
        numeric_part = x_str.replace("patient_", "")
    else:
        numeric_part = x_str
    # Pad numeric part to 5 digits
    numeric_part = numeric_part.zfill(5)
    return f"patient_{numeric_part}"

df["patient_id"] = df["patient_id"].apply(format_patient_id)

# -----------------------------
# Rename column from patient_id → patient_
# -----------------------------
df = df.rename(columns={"patient_id": "patient_"})

# -----------------------------
# Save the updated dataset
# -----------------------------
df.to_csv(OUTPUT_CSV, index=False)

print(f"✅ Updated dataset saved to '{OUTPUT_CSV}'")
```
